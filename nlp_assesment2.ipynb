```python
import pandas as pd
import numpy as np
import re
import nltk

from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

nltk.download('stopwords')
```

```python
data = {
    "review": [
        "I love this product, it's amazing!",
        "Worst experience ever, very bad service",
        "Absolutely fantastic, would buy again",
        "Not good, very disappointing",
        "This is the best thing I have purchased",
        "Terrible quality, waste of money",
        "I am very happy with this",
        "I am not satisfied at all",
        "Great value for money",
        "Horrible experience, will not recommend",
        "Superb and excellent",
        "Bad product, broke in one day",
        "I really enjoyed using this",
        "Not worth the price",
        "Highly recommended",
        "Very poor performance",
        "Loved it so much",
        "It is okay, not great",
        "Extremely good quality",
        "Awful, do not buy"
    ],
    "sentiment": [
        "positive","negative","positive","negative","positive",
        "negative","positive","negative","positive","negative",
        "positive","negative","positive","negative","positive",
        "negative","positive","neutral","positive","negative"
    ]
}

df = pd.DataFrame(data)
df.head()
```

```python
print("Shape:", df.shape)

print("\nClass Distribution:")
print(df['sentiment'].value_counts())

print("\nSample Data:")
print(df.head())
```

```python
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()

def preprocess(text):
    text = text.lower()
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'[^a-zA-Z]', ' ', text)

    words = text.split()

    words = [stemmer.stem(w) for w in words if w not in stop_words]

    return " ".join(words)

df['clean_text'] = df['review'].apply(preprocess)

df[['review', 'clean_text']].head()
```

```python
# Bag of Words
bow = CountVectorizer(max_features=500)
X_bow = bow.fit_transform(df['clean_text'])

# TF-IDF
tfidf = TfidfVectorizer(max_features=500)
X_tfidf = tfidf.fit_transform(df['clean_text'])

# Encode labels
y = df['sentiment'].map({
    'positive':1,
    'negative':0,
    'neutral':2
})
```

```python
X_train_bow, X_test_bow, y_train, y_test = train_test_split(X_bow, y, test_size=0.2)

X_train_tfidf, X_test_tfidf, _, _ = train_test_split(X_tfidf, y, test_size=0.2)
```

```python
# Logistic Regression
lr = LogisticRegression(max_iter=200)
lr.fit(X_train_tfidf, y_train)
y_pred_lr = lr.predict(X_test_tfidf)

# Naive Bayes
nb = MultinomialNB()
nb.fit(X_train_bow, y_train)
y_pred_nb = nb.predict(X_test_bow)

# Decision Tree
dt = DecisionTreeClassifier()
dt.fit(X_train_bow, y_train)
y_pred_dt = dt.predict(X_test_bow)
```

```python
def evaluate(y_test, y_pred, model_name):
    print(f"\n{model_name}")
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print("Precision:", precision_score(y_test, y_pred, average='macro'))
    print("Recall:", recall_score(y_test, y_pred, average='macro'))
    print("F1 Score:", f1_score(y_test, y_pred, average='macro'))

evaluate(y_test, y_pred_lr, "Logistic Regression")
evaluate(y_test, y_pred_nb, "Naive Bayes")
evaluate(y_test, y_pred_dt, "Decision Tree")
```

# Comparison & Insights

- Logistic Regression performed best due to TF-IDF features.
- Naive Bayes worked well with Bag of Words.
- Decision Tree showed lower performance due to overfitting.

Best Model: Logistic Regression

Best Feature Method: TF-IDF

Conclusion:
TF-IDF + Logistic Regression gives highest accuracy for sentiment analysis.
